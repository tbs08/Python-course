```python
%matplotlib notebook
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

# note that you won't be able to render the map if this script didn't run in the online platform via coursera
def leaflet_plot_stations(binsize, hashid):
#  since the file is online but i changed the 
    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'997b94c98cc130812ea657b0bf1149e354d28ba15c72d9e4b28fb07f')
```

```python
%matplotlib notebook
import pandas as pd
import numpy as np
df=pd.read_csv("statecollege.csv")

#======= below is to remove any data on 2/29========
df['Month']=[d.split('/')[0] for d in df.Date]
df['Day']=[d.split('/')[1] for d in df.Date]
df['Year']=[d.split('/')[2] for d in df.Date]
# type(df['Year']) is string
df['MonthDay']=df['Month']+"/"+df['Day']
df=df.drop(df[df.MonthDay=='02/29'].index)


# =====below is to convert to celsius and F and extract 2005-2014 data and 2015 data====

df['C']=df['Data_Value']/10
df['F']=df['C']*1.8+32
df1=df[(df['Year']<'15') & (df['Year']>'04')]
df2=df[df['Year']=='15']


# =====below is to find the historical high and low for each day over the years of 2005 to 2014 and 2015 record-breaking high and low===
max=df1[df1['Element']=="TMAX"]
high=max.set_index('MonthDay').groupby(level=0)['C','F'].max()


min=df1[df1['Element']=='TMIN']
low=min.set_index('MonthDay').groupby(level=0)['C','F'].min()

max15=df2[df2['Element']=="TMAX"]
high15=max15.set_index('MonthDay').groupby(level=0)['C','F'].max()
high15.columns=['C15','F15']

min15=df2[df2['Element']=='TMIN']
low15=min15.set_index('MonthDay').groupby(level=0)['C','F'].min()
low15.columns=['C15','F15']

# ======convert date string to date format========
# change the index from string to date using to_datetime, the previous date string was 08/28, however, if it's 8/28, the conversion should also work

high.rename(index=lambda x: pd.to_datetime(x, format="%m/%d"),inplace=True )
low.rename(index=lambda x: pd.to_datetime(x, format="%m/%d"),inplace=True )
low=low.sort_index()
high=high.sort_index()

high15.rename(index=lambda x: pd.to_datetime(x, format="%m/%d"),inplace=True )
low15.rename(index=lambda x: pd.to_datetime(x, format="%m/%d"),inplace=True )
high15=high15.sort_index()
low15=low15.sort_index()

# ====below is to find the record-breaking points in 2015====
h=pd.merge(high,high15,how="inner",left_index=True, right_index=True)
h['break']=h['C15']-h['C']
hb=h[h['break']>0].sort_index()

l=pd.merge(low,low15,how="inner",left_index=True, right_index=True)
l['break']=l['C15']-l['C']
lb=l[l['break']<0].sort_index()

```

```python
# ===let's begin line plot and overlay scatterplot====
# import necessary modules
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
rc('mathtext',default='regular')
import matplotlib.dates as mdates


# create a figure instance first

plt.figure()

# although line plot doesn't require x axis but we can still plot x axis, in this case, it's important


plt.plot(low.index,low['C'].values,'-',label='low (2005-2014)',linewidth=1)
plt.plot(high.index,high['C'].values,'-',label='high (2005-2014)',linewidth=1)
plt.scatter(hb.index,hb['C'].values,s=10,label='2015 high break points',c='red')
plt.scatter(lb.index,lb['C'].values,s=10,label='2015 low break points',c='blue')


# ======below start specifying the ticks and labels===
ax=plt.gca()

# first, convert days of the year to month for better appearance. 
myfmt=mdates.DateFormatter('%b')
months=mdates.MonthLocator()
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(myfmt)

# add secondary axis using twinx()
ax1=ax.twinx()

# set up the two axes
ax.set_yticks(np.arange(-30,50,10))
ax.set_yticklabels(str(i)+'$^\circ$C' for i in np.arange(-30,50,10))
    
ax1.set_yticks(np.arange(-30,50,10))
ax1.set_yticklabels(str(int(i*1.8+32))+'$^\circ$F' for i in np.arange(-30,50,10))

# set legend and title
ax.legend(loc=4,prop={'size':6})
ax.set_title('State Colelge Daily Historical High and Low Temperature over 2005-2014 \n and 2015 Record-breaking Points',fontsize=8)

ax.fill_between(high.index,high['C'].values, low['C'].values, facecolor='lightslategrey',alpha=0.25)

```
